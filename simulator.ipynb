### GPS -> World

```python
x_world = np.zeros([100,3])
for i in range(x_gps.shape[0]):
    x_world[i] = coordinate.gps2world(x_gps[i])
```

```python
'''
@function: draw the 3d plot of dataset
@param: dataset - an array of x,y,z points
@return: a 3D plot of points
'''
def plot_3d(dataset):

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    xs = dataset[:,0]
    ys = dataset[:,1]
    zs = dataset[:,2]
    ax.scatter(xs, ys, zs, marker='x')

    ax.set_xlabel('North')
    ax.set_ylabel('East')
    ax.set_zlabel('Latitude')

    plt.show()
```

```python
plot_3d(x_gps)
```

```python
plot_3d(x_world)
```
